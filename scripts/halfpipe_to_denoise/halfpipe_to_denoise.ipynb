```python

import pandas as pd
import numpy as np
```

```python


# Set your file path here
file_path = "/home/seann/scratch/denoise/fmriprep-denoise-benchmark/giga_timeseries/brain_canada_fmriprep-20.2.7lts_1732217118/fmriprep-20.2.7lts/atlas-schaefer7networks/sub-01/sub-01_task-rest_space-MNI152NLin2009cAsym_atlas-schaefer7networks_nroi-100_desc-baseline_timeseries.tsv"

# Try reading the file without assuming a header
df = pd.read_csv(file_path, delimiter="\t", header=None)

# Function to check if the first row is numeric (i.e., no header) or not.
def is_header_present(df):
    try:
        pd.to_numeric(df.iloc[0])
        return False  # Conversion succeeded: first row is data
    except Exception:
        return True   # Conversion failed: first row is header

# Determine if the file already has a header
header_present = is_header_present(df)

print("=== File Information ===")
print(f"File path: {file_path}")
print(f"Shape (rows x columns): {df.shape}\n")

if header_present:
    print("Header row detected; using the first row as column names.\n")
else:
    print("No header row detected (first row is numeric data).\n")
    
print("=== First 5 Rows ===")
print(df.head(5))
print("\n=== Last 5 Rows ===")
print(df.tail(5))
print("\n")

# If no header is present, assign temporary column names for further inspection
if not header_present:
    temp_cols = [f"col_{i}" for i in range(df.shape[1])]
    df.columns = temp_cols

print("=== Column Data Types ===")
print(df.dtypes)
print("\n=== Summary Statistics ===")
# Try converting all columns to numeric for summary stats
df_numeric = df.apply(pd.to_numeric, errors='coerce')
print(df_numeric.describe())
print("\n=== Missing Values per Column ===")
print(df_numeric.isna().sum())
```
